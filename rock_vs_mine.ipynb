split data(feature). and target

```python
x=soner.drop(columns='R',axis=1)
y=soner['R']
```

```python
print(x,y)
```

stratify -> rock mine terget basis a predict

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

**when dtat range under 0-1 thn no need to standardization**

```python
x_train.std()
```

**model trainnig using logistic regression**

```python
model=LogisticRegression()
```

training dtaa basic a fit hoba not test data

```python
model.fit(x_train,y_train)
```

Accuressy on trainnng *data*

```python
x_train_predction=model.predict(x_train)
trainning_data_acuacy=accuracy_score(x_train_predction,y_train)
```

```python
print('acuuracy on trainning data',trainning_data_acuacy)
```

 Accuressy on trainnng *data*section

```python
x_test_predction=model.predict(x_test)
x_test_accuracy=accuracy_score(x_test_predction,y_test)
```

```python
print(x_test_accuracy)
```
